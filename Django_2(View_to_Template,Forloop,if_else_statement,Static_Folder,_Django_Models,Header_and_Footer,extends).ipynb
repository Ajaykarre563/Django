# **if-else statement**

**In Django templates, the if-else statement is used to add logic to your HTML pages. It allows you to display different content based on certain conditions, just like in regular programming.**

# **Syntax:**



```
{% if condition %}
   <!-- code if condition is true -->
{% else %}
   <!-- code if condition is false -->
{% endif %}

```

# **Header and Footer in Django HTML Template (Fix Header & Footer)**

# **Include:**

# 1. **{% include %}**

The {% include %} tag is used to insert the contents of another template into your current template. It's like embedding a partial HTML file inside your main HTML file.

# **When to Use {% include %}:**
When you want to reuse a small part of a template across multiple pages (like a header, footer, sidebar, or any repeating section).

Its perfect for static content that doesnt change per page.

Example:
Lets say you have a header and footer that you want to include in multiple pages.

**Instead of repeating header and footer code on every page, Django allows you to reuse components with:**



```
{% include "header.html" %}
{% include "footer.html" %}

```



# ** Why this is useful:**
Keeps code DRY (Don't Repeat Yourself)

Easy to update one file, and changes reflect on all pages

Ideal for navigation menus, logos, and footers

# **Extends:**

# **2. {% extends %}**

The {% extends %} tag is used to create a base template that other templates can inherit from. It's part of template inheritance in Django, where a template can extend or inherit from another base template, allowing you to define a common structure (like a header, footer, sidebar) in one place and have all other pages reuse that structure.

# **When to Use {% extends %}:**

When you want to define a base layout for your site that can be used across multiple pages (e.g., header, footer, navigation).

It allows you to override specific parts of the layout (like content areas) on different pages.
