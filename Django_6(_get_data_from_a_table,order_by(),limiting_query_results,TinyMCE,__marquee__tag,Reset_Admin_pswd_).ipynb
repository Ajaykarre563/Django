# **Implementation Steps for Custom Template Filter:**

Create a templatetags directory inside your Django app (same level as models.py).

Inside templatetags, create a .py file (e.g., custom_filters.py).

Register your custom filter using @register.filter.

# **📦 Useful Built-In Python Functions in Custom Filters**

When writing your own filters, you often use:

# **Function	Purpose**

slice, list slicing	Extract sub-parts from strings/lists

datetime	Date/time formatting or calculations

html.escape()	Escape HTML (optional from html module)

# **What is mean by marquee tag?**

The marquee tag in HTML is used to create a scrolling text or image (usually horizontally across the screen). Though it's outdated and not recommended in modern web design (as it's not supported in HTML5), it still works in most browsers for simple use cases like displaying scrolling news.

# **what is mean by nbsp;(non-breaking space)?**

# **🔍 What it does:**
It adds a space between elements or text, but unlike a regular space (pressed from the keyboard), it won't break onto a new line even if the text wraps.

# **📌 Use Cases:**

To insert multiple spaces (since normal HTML collapses whitespace).

To keep words together without breaking on line wrap.
