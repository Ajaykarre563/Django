# **Manual form validation in Django**

**Manual form validation in Django is the process of writing custom logic to validate form input without relying solely on Djangos built-in Form or ModelForm validation methods. This gives developers full control over how data is checked before being processed or saved.**



# **Note:**
Manual Form Validation in Django means that the data is checked before any processing (like saving to the database or performing calculations).
This ensures only clean, valid input is used in your application logic.

** Flow Recap:**

User submits form 

request.POST is read 

Each field is manually validated 

If all fields are valid  process data

If errors exist  send them back to the user

# **What is a Model in Django**

**In Django, a Model is a Python class that defines the structure of your database table. It acts as a bridge between your Python code and the underlying database.**

# ** Key Points:**
A model is a class that subclasses django.db.models.Model.

Each class attribute represents a database field (e.g., CharField, IntegerField).

Django automatically creates the SQL table from this model.

You can query, insert, update, and delete data using Django's ORM.
