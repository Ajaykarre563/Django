# **âœ… What is Redirect in Django?**
Redirect in Django is a way to send the user from one URL to another. It is commonly used after form submission, login, logout, or any event where you want to show a different page.



# **âœ… 1. urls.py**



```
from django.urls import path
from . import views

urlpatterns = [
    path('', views.home, name='home'),
    path('login/', views.login_view, name='login'),
    path('dashboard/', views.dashboard, name='dashboard'),
]

```



# **âœ… 2. views.py**



```
from django.shortcuts import render, redirect

def home(request):
    return render(request, 'home.html')

def login_view(request):
    # Simulating a login process
    if request.method == 'POST':
        # In real use, you'd check user credentials here
        return redirect('dashboard')  # Redirect after login
    return render(request, 'login.html')

def dashboard(request):
    return render(request, 'dashboard.html')

```

**In Django, the HTML action URL in a form tag specifies the endpoint (URL) to which form data will be submitted when the form is submitted.**

# **Example:**

# **In views.py:**



```
from django.shortcuts import render
from django.http import HttpResponse

def submit_form(request):
    if request.method == "POST":
        name = request.POST.get('name')
        return HttpResponse(f"Hello, {name}!")
    return render(request, 'form.html')

```



# **In urls.py:**



```
path('submit/', views.submit_form, name='submit_form')

```



# **In form.html:**



```
<form method="post" action="{% url 'submit_form' %}">
  {% csrf_token %}
  <input type="text" name="name" placeholder="Enter your name">
  <button type="submit">Submit</button>
</form>

```



# **Expected I/p and O/p**

# **building a basic calculator in Django**

# **Step 1: Create a Django Project**



```
django-admin startproject calculator_project
cd calculator_project

```



# **Step 2: Create an App**



```
python manage.py startapp calculator
```



# **Step 3: Register the App in settings.py**
In calculator_project/settings.py, add 'calculator' to INSTALLED_APPS.



```
INSTALLED_APPS = [
    ...
    'calculator',
]

```



# **Step 4: Create the HTML Form**
Inside your app folder (calculator), create a folder named templates, and inside that, create a file named calculator.html:



```
<!-- calculator/templates/calculator.html -->
<h2>Simple Django Calculator</h2>
<form method="post">
    {% csrf_token %}
    <input type="number" name="num1" required>
    <select name="operation">
        <option value="add">+</option>
        <option value="subtract">âˆ’</option>
        <option value="multiply">Ã—</option>
        <option value="divide">Ã·</option>
    </select>
    <input type="number" name="num2" required>
    <button type="submit">Calculate</button>
</form>

{% if result is not None %}
    <h3>Result: {{ result }}</h3>
{% endif %}

```



# **Step 5: Create Views in views.py**



```
# calculator/views.py
from django.shortcuts import render

def calculator_view(request):
    result = None
    if request.method == 'POST':
        num1 = float(request.POST['num1'])
        num2 = float(request.POST['num2'])
        operation = request.POST['operation']

        if operation == 'add':
            result = num1 + num2
        elif operation == 'subtract':
            result = num1 - num2
        elif operation == 'multiply':
            result = num1 * num2
        elif operation == 'divide':
            result = num1 / num2 if num2 != 0 else "Cannot divide by zero"

    return render(request, 'calculator.html', {'result': result})

```

# **Create a Django Form**
Create a file: forms.py



```
from django import forms

class NumberForm(forms.Form):
    number = forms.IntegerField(label='Enter a number')

```



# **Create the View**
Update views.py:



```
from django.shortcuts import render
from .forms import NumberForm

def check_even_odd(request):
    result = None
    if request.method == 'POST':
        form = NumberForm(request.POST)
        if form.is_valid():
            num = form.cleaned_data['number']
            result = f"{num} is Even" if num % 2 == 0 else f"{num} is Odd"
    else:
        form = NumberForm()
    return render(request, 'check.html', {'form': form, 'result': result})

```



# **Create Template**
Create file: templates/check.html



```
<!DOCTYPE html>
<html>
<head>
    <title>Even or Odd Checker</title>
</head>
<body>
    <h2>Check if a Number is Even or Odd</h2>
    <form method="post">
        {% csrf_token %}
        {{ form.as_p }}
        <button type="submit">Check</button>
    </form>
    {% if result %}
        <h3>{{ result }}</h3>
    {% endif %}
</body>
</html>

```



# **Set the URL**
In urls.py of your app:



```
from django.urls import path
from . import views

urlpatterns = [
    path('', views.check_even_odd, name='check_even_odd'),
]

```
