# **what is mean by  URLs?**

**A URL (route) in Django is like giving an address to a specific page on your website.**

When a visitor types a web address (like https://example.com/blog/) into their browser:

Django checks if that address (URL) is written in your code.

If yes, it will run a function (called a "view") that builds the page and shows it to the user.



# **Basic Concepts:**

URL: The address typed in the browser.

Example: https://www.wscubetech.com/blog/

Route: Part of the URL after the domain.

In the example above, /blog/ is the route.

View Function: A Python function that takes a request and returns a response (like an HTML page).



# **Django URL Mapping Example:**



```
# urls.py

from django.urls import path
from . import views

urlpatterns = [
    path('', views.homepage),              # https://example.com/ ‚Üí homepage
    path('blog/', views.blog_list),        # https://example.com/blog/ ‚Üí list of blog posts
    path('blog/<int:id>/', views.blog_post) # https://example.com/blog/3/ ‚Üí specific blog post
]

```



# **what is mean by Views?**

Views are Python functions (or classes) that take a request and return a response (like an HTML page, JSON, etc.).

# **üì¶ How it works:**

# **When someone visits a URL:**

Django checks your urls.py to see which view is connected to that URL.

It runs that view (a function or class).

That view prepares the data and returns the page the user should see.



# **Example: Function-Based View**



```
from django.http import HttpResponse

def homepage(request):
    return HttpResponse("Welcome to the homepage!")

```

# **what is mean by  dynamic URl?**

**A dynamic URL in Django is a URL that changes based on user input, like an ID or name, and shows different content.**

# **Example of Static URL:**



```
path('course/', views.course)

```
This always goes to the same page ‚Äî like /course/.



# **Example of Dynamic URL:**



```
path('course/<int:id>/', views.course)

```
This changes based on the id, like:

/course/1/

/course/5/

In the view, you can access the id:



```
def course(request, id):
    return HttpResponse(f"You selected course ID: {id}")

```
